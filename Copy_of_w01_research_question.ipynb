## 2. The question: decision, action, cost of a wrong call

```python

Decision: Which website pages should be prioritized for content refresh?
Action: The SEO or content team can review and update the highest-priority pages first.

Cost of a Wrong Recommendation:
- A declining page may be missed, resulting in continued traffic loss.
- Time and effort may be spent updating pages that did not need immediate attention.

Machine learning can support this decision by analyzing multiple page metrics together and producing a ranked list of pages for review.
```
